# 單維度數據分析

```python
var_list = ['Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms', 'Views', 'Likes', 'Comments', 'Stream']
```

```python
# Plot box plots for each numerical feature and save as images
for feature in var_list:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df_data[feature])
    plt.title(feature)
    plt.xlabel(feature)
    # plt.gca().invert_yaxis()  # equivalent to coord_flip()
    plt.savefig(f'plot_result/{feature}.png')
    plt.show()
```

## 相關性分析

```python
var_list = ['Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms', 'Views', 'Likes', 'Comments', 'Stream']
```

```python
df_data_numeric_var = df_data[var_list]
```

```python
df_data_numeric_var
```

```python
df_data_corr = df_data_numeric_var.corr()
```

```python
import seaborn as sns

cmap = sns.diverging_palette(230, 20, as_cmap=True)
plt.figure(figsize=(12, 10))
sns.heatmap(df_data_corr, annot=True, 
                xticklabels=df_data_corr.columns, 
                yticklabels=df_data_corr.columns,
                cmap = cmap,
                vmin=-1, vmax=1)

plt.show()
```

```python
import statsmodels.api as sm
```

### View vs. Stream

```python
# View vs. Stream
plt.figure(figsize=(8, 6))
sns.regplot(x='Stream', y='Views', data=df_data, line_kws={"color": "black"}, ci=None)
plt.title('Stream vs. Views')
plt.savefig('plot_result/Stream_View.png')
plt.show()

df_data_plot = df_data.loc[:, ['Views','Stream']].dropna()
# Fit linear model: Stream ~ Views
X = df_data_plot['Views']
y = df_data_plot['Stream']
X = sm.add_constant(X)  # add intercept
model = sm.OLS(y, X).fit()
print(model.summary())  # R-squared: 0.3631
```


### Likes vs. Comments

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Likes', y='Comments', data=df_data, line_kws={"color": "black"}, ci=None)
plt.title('Likes vs. Comments')
plt.savefig('plot_result/Likes_Comments.png')
plt.show()

# Fit linear model: Comments ~ Likes
df_data_plot = df_data.loc[:, ['Likes','Comments']].dropna()
X = df_data_plot['Likes']
y = df_data_plot['Comments']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())  # R-squared: 0.3988
```


### Likes vs. Views

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='Likes', y='Views', data=df_data, line_kws={"color": "black"}, ci=None)
plt.title('Likes vs. Views')
plt.savefig('Likes_Views.png')
plt.show()

# Fit linear model: Views ~ Likes
df_data_plot = df_data.loc[:, ['Likes','Views']].dropna()
X = df_data_plot['Likes']
y = df_data_plot['Views']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())  # R-squared: 0.7958
```
